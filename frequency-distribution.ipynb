### Frequency Distribution üóíÔ∏è

Frequency distribution is a way to visualize how data from a certain variable is distributed. It is one among several alternatives (such as histograms, for example) to visualize possible concentrations of data in a specific set or even concentrations of data individually by analyzing each record (although this is less common).

It goes beyond simply counting how many times a particular record appears - It is a fundamental concept for decision-making, detecting possible trends in a specific group of data, and so on... Let's explore more about this in the following text:

#### Why do I need to know how my variable's data is distributed?

Imagine a scenario where you need to create a machine learning algorithm that classifies whether a person is likely to have heart failure based on historical data.

In addition to understanding possible relationships, it's also important to identify potential data concentrations by comparing them with our output variable. For example, understanding if there is a higher concentration of people of a certain age or age group among patients who have heart failure.

There are different ways to visualize data distribution, and one of them is **frequency distribution**. Let's delve into it in more detail:

* face: The sample value I want to visualize.

* fi: Known as absolute frequency or simple frequency, it's the number of times the Face is displayed.

* fr: It's the percentage that fi represents in relation to the sum of frequencies. The formula would be:

  fri = fi/n (where n is the sum of fi)


#### Frequency distribution with class

   Frequency distribution with classes is **generally** applied to continuous quantitative variables. This is because the concept of a continuous quantitative variable involves different values that fluctuate within a certain range, potentially infinitely. Therefore, do you agree that creating a simple frequency distribution table for a variable with large variations in its values is impractical? For example: Imagine instead of analyzing the 25 patients from our fictitious table above, we had millions of records. Therefore, we divide our data into groups consisting of intervals between 2 values.



 Frequency distribution with classes involves creating classes that group the data from our variable. The idea is to analyze the frequency in a broader and less individual manner.


#### Separating the values of our variable into classes


<br/>

> ##### How many classes should we have? 
 
 To separate our values into classes, we can adopt one of two strategies. One is to eyeball it üëÅÔ∏è, meaning after extracting the values, I apply a range that I find appropriate and move forward. The other strategy involves applying some techniques. It's not about right or wrong here; it's about having different alternatives for a process. However, in most cases, following certain mathematical techniques yields more effective results.

Let's proceed with the techniques here so that you have an example. The first one is knowing how many classes I should divide my data into, and for that, we can apply the `Sturges Formula`:

~~~python
     ‚Äãk = 1 + 3.3 . log(n)
~~~ 


 `n`: It's the total number of records in the sample - In our case, it's 25.


Resolvendo a f√≥rmula temos:  5,61 - (Arredondando) Oodemos afirmar que segundo a `F√≥rmula de Sturges` devemos ter 6 classes


<br/>

> ##### What should be the interval between each class?

 Ok, at this point, we know we should have 6 classes, but what should be the interval between their values? The question I want to answer here is: If I have 6 classes, what should be the starting and ending values for each one to determine their intervals? There's also a technique we can apply for this, let's see the formula:


~~~python
    Ai = At / k
~~~ 

<br/>

* Ai: Interval amplitude

* At: Total amplitude = Largest number in the range - Smallest number in the range

* k: Number of classes


Solving the formula, we have:

Smallest age is 40 and largest age is 70.

Ai = (70 - 40) / 6

The result of the calculation above is 5, so we should have 6 classes where each interval spans 5 years. Our frequency distribution table using classes looks like this:


| classes | fi | fr   |
|---------|----|------|
| 40 ‚îú 45 | 3  | 0.12 |
| 45 ‚îú 50 | 5  | 0.2  |
| 50 ‚îú 55 | 4  | 0.16 |
| 55 ‚îú 60 | 7  | 0.28 |
| 60 ‚îú 65 | 3  | 0.12 |
| 65 ‚îú 70 | 2  | 0.08 |
| 70 ‚îú 75 | 1  | 0.04 |


> The symbol ‚îú indicates that the value on the left is included in the interval while the value on the right is not, read as:

~~~
inclusive ‚îú exclusive
~~~

     
Notice that we ended up with 7 classes instead of the 6 established by the Sturges' Formula. This is possible in very small datasets because the total number of records nn, which determines the number of classes, may not be sufficient to precisely determine the total number of classes. In this case, we added an additional class from 70 to 75.

With the boxplot below, it's possible to see that a value above this last interval, between 70 ‚îú 75, is considered an outlier. This might lead us to conclude that 7 is a more suitable number for the total number of classes.

```python
import pandas as pd 
import numpy as np

FAKE_DATA = 76 

data = np.array([55,62, 45, FAKE_DATA, 50, 58, 48, 65, 52, 60, 42, 68, 55, 57, 47, 50, 55, 40, 63, 45, 59, 53, 58, 48, 42])

df = pd.DataFrame(data)

df.boxplot(grid=False);
```
